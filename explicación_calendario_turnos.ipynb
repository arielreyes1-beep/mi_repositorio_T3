#Funcionamiento calendario
El segundo trabajo realizado en el ramo de lenguaje de programación fue generar un calendario para planificar turnos.
Para la correcta explicación del código, se organizó la explicación de la siguiente forma:
* **Importación de las bibliotecas**: El primer paso fue importar las bibliotecas que se utilizarán, estas fueron:
  1. **tkinter**: Para crear la ventana en la cual se mostrará el calendario planificador.
  2. **calendar**: Utilizada para trabajar con calendarios.
  3. **datetime**: Utilizada para manejar fechas.

* **configuración de ventana, interfaz e inicializacion de variables**: Se comenzó configurando la ventana; además se crearon los frames donde se encontrará la interfaz del planificador, ubicandolos según su fila y columna.
Por otro lado se definieron e inicializaron las variables a utilzar.

* **Definición de funciones**: Con las variables inicializadas, se definieron las funciones que utilizará el planificador, donde se utilizaron las siguientes:

1. Mostrar el calendario: Mediante un bucle *for* se verifica si hay un calendario en pantalla, en el caso de existir uno, este se borra.
Posteriormente, se crea un calendario con sus respectivos meses y días (para estos se uso un bucle *for*).
Luego se obtienen los días de cada mes, los cuales se organizan en listas y son agregados al calendario dandole a cada uno la característica de ser botones
Finalmente en caso de haber un turno, este se aplica en todo el calendario.

2. Aplicar turnos: Se comienza revisando si hay turnos ya definidos, luego se calcula el ciclo de turnos y el dia del año donde comienza el mes, Finalmente recorre cada día para decidir si es un día de trabajo o descanso asignando un color según corresponda.

3. Turnos mensuales: Se definen los días de trabajo y descanso para crear un turno, convierte el día del mes en día del año y se dibuja nuevamente el calendario para mostrar los colores asignados.

4. Seleccionar día: Se registran los días que son seleccionados mediante los botones, en el caso de haber 2 seleccionados se llama a la función *turnos mensuales* y se borran los días seleccionados con esta función.

5. Funciones para botones fuera del calendario: Se crean funciones para mes siguiente y anterior considerando el cambio de año cuando corresponda.
Además se crea una función de limpiar el calendario para eliminar los colores en caso de haber y mostrar el planificador limpio.

* **botones para cambiar meses y limpiar calendario**: Se crean y posicionan los botones para las funciones previamente definidas y se le asigna a cada uno su acción correspondiente.
