```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')

regiones = ["Noroccidental", "Nororiental", "Central", "Meridional", "Insular"]
poblacion = [15923805, 11597719, 12972742, 12099201, 6340710]
colores = ["#C46200", "#E03C31", "#007FA3", "#7F00FF", "#FFD700"]

x = np.arange(len(regiones))

fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(x, poblacion, width=0.9, color=colores, edgecolor="white", linewidth=0.7)

for i, valor in enumerate(poblacion):
    ax.text(i, valor + 200000, f"{valor:,}".replace(",", "."),
            ha='center', va='bottom', fontsize=9)

ax.set(
    xlim=(-0.5, len(regiones) - 0.5),
    ylim=(0, max(poblacion) * 1.15),
    xticks=x,
    xticklabels=regiones,
)

plt.xticks(rotation=20)
plt.title("Poblacin de Italia por regin geopoltica")
plt.xlabel("Regin geopoltica")
plt.ylabel("Poblacin")
plt.tight_layout()
plt.show()
```
