#Funcionamiento calculadora
El primer trabajo realizado en el ramo de lenguaje de programación fue generar una calculadora simple y que presentara alguna función fuera de lo normal o entretenido.
Para la correcta explicación del código, se organizó la explicación de la siguiente forma:
* **Importación de la biblioteca tkinter**: El primer paso fue importar la biblioteca tkinter, para crear la ventana en la cual se mostrará la calculadora.

* **Inicialización de variables y primera configuración de la ventana**: En esta sección se crearon diferentes variables como por ejemplo la de los números **A** y **B** para inicializarlas dentro del programa y utilizarlas posteriormente en diferentes funciones. Además se configuró la ventana y un espacio donde se colocará la interfaz de la calculadora.

* **Definición de funciones**: Con las variables inicializadas, se definieron las funciones que tendrá la calculadora, donde finalmente se utilizaron las siguientes:
    1. Mostrar los números en pantalla: Para esto se editó el texto que se mostraba en la pantalla, utilizando los valores el numero presionado y transformándolo en texto.
    2. Seleccionar la operación: Se utilizaron las variables globales de número **A** y operación, donde el número ingresado mediante los botones se transforma en el número **A** y el botón presionado con una operación la selecciona y limpia la pantalla de la calculadora.
    3. Calcular el resultado: Se llaman a las variables globales necesarias y se utilizan los diferentes condicionales para definir la operación que la calculadora realizará, para posteriormente mostrar en la pantalla el resultado. Además en esta función se encuentra una característica la cual, mediante un condicional *if*, indica que la calculadora mostrará en pantalla una imagen cuando el resultado de una operación sea *2*.
    4. Borrar: Se eliminan los números en pantalla y en el caso de que la imágen, esta también se elimina.

* **Configuración de la interfaz de la calculadora**: Finalmente, se creó la interfaz completa de la calculadora, utilizando listas para los botones numéricos y un bucle *for* para recorrerlas, por otro lado, los botones de operaciones y algunos extras se colocaron manualmente indicando su fila y columna respectiva.
